# **All QR code with Tittle**

```python
import qrcode
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display
import os

def generate_qr_with_title(url, title, output_file, font_path=None, box_size=10, border=5, fill_color="black", back_color="white", font_size=30):
    """
    Generates a QR code with a title below it and saves it as an image.

    Parameters:
        url (str): The URL to embed in the QR code.
        title (str): The title text to add below the QR code.
        output_file (str): The file path to save the generated QR code image.
        font_path (str): Path to a .ttf font file. Defaults to system font if None.
        box_size (int): Size of the QR code boxes.
        border (int): Border size of the QR code.
        fill_color (str): Color of the QR code.
        back_color (str): Background color of the QR code.
        font_size (int): Font size for the title text.

    Returns:
        PIL.Image.Image: The generated QR code image with the title.
    """
    # Create QR code
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_H,
        box_size=box_size,
        border=border,
    )
    qr.add_data(url)
    qr.make(fit=True)

    # Generate QR code image
    qr_image = qr.make_image(fill_color=fill_color, back_color=back_color).convert('RGB')

    # Load font
    if font_path is None:
        font_path = "/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf"  # Default font path for Linux
    if not os.path.exists(font_path):
        raise FileNotFoundError(f"Font file not found: {font_path}")

    font = ImageFont.truetype(font_path, font_size)
    draw = ImageDraw.Draw(qr_image)

    # Calculate text position
    text_bbox = draw.textbbox((0, 0), title, font=font)
    text_width = text_bbox[2] - text_bbox[0]
    text_height = text_bbox[3] - text_bbox[1]
    img_width, img_height = qr_image.size
    text_position = ((img_width - text_width) // 2, img_height - text_height - 10)  # Adjust padding as needed

    # Add title to the image
    draw.text(text_position, title, font=font, fill=fill_color)

    # Save the image
    qr_image.save(output_file)
    print(f"QR Code with title saved to {output_file}")

    return qr_image


# Define your platforms and titles
platforms = {
    "YouTube": ("https://www.youtube.com/@content123", "Subscribe Content on Demand"),
    "Facebook": ("https://www.facebook.com/content.on.demand.2024", "Follow us on Facebook"),
    "LinkedIn": ("https://www.linkedin.com/in/content-on-demand-917146269", "Connect on LinkedIn"),
    "Github": ("https://github.com/Nazmul0005", "Explore our Github Repository"),
    "Instagram": ("https://www.instagram.com/contentondemand904?utm_source=qr&igsh=MTlneGt6YzIwM2R3Mw==", "Follow us on Instagram"),
    "Threads": ("https://www.threads.net/@contentondemand904", "Join us on Threads"),
    "X (Twitter)": ("https://x.com/content904?t=PaAsW4JHtSFOxTrErqrrtw&s=09", "Follow us on X (Twitter)"),
    "TikTok": ("https://www.tiktok.com/@contentondemand?_t=8rYWYQefOPi&_r=1", "Watch us on TikTok"),
}

# Generate and display QR codes
for platform, (url, title) in platforms.items():
    output_file = f"{platform.lower()}.png"
    qr_image = generate_qr_with_title(
        url=url,
        title=title,
        output_file=output_file,
        font_path="/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf",  # Adjust for your system
        box_size=10,
        border=4,
        fill_color="black",
        back_color="white",
        font_size=24,
    )
    print(f"Displayed QR code for {platform}.\n")
    print()
    print()
    display(qr_image)
```

# **QR Code with Logo**

```python
import qrcode
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display
import os

def generate_gradient_image(width, height, color_start, color_end):
    """Generates a gradient image from color_start to color_end."""
    gradient = Image.new('RGB', (width, height), color_start)
    for y in range(height):
        r = int(color_start[0] + (color_end[0] - color_start[0]) * (y / height))
        g = int(color_start[1] + (color_end[1] - color_start[1]) * (y / height))
        b = int(color_start[2] + (color_end[2] - color_start[2]) * (y / height))
        ImageDraw.Draw(gradient).line((0, y, width, y), fill=(r, g, b))
    return gradient

def generate_qr_with_title(url, title, output_file, font_path=None, box_size=10, border=5, fill_color="black", back_color="white", font_size=30, logo=None, gradient=False):
    """
    Generates a QR code with a title below it, a logo, and optionally a gradient background.
    """
    # Create QR code with higher error correction for logo placement
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_Q,  # Higher error correction
        box_size=box_size,
        border=border,
    )
    qr.add_data(url)
    qr.make(fit=True)

    # Create base QR code image
    qr_image = qr.make_image(fill_color=fill_color, back_color=back_color).convert('RGB')

    # Apply gradient if enabled
    if gradient:
        gradient_image = generate_gradient_image(qr_image.width, qr_image.height, (255, 0, 0), (0, 0, 255))  # Red to Blue gradient
        qr_image = Image.blend(qr_image, gradient_image, 0.5)

    # Add logo to the center (optional)
    if logo:
        logo = Image.open(logo).convert("RGBA")
        logo_width, logo_height = logo.size
        qr_width, qr_height = qr_image.size

        # Resize the logo if it's too large
        max_logo_size = min(qr_width, qr_height) // 4  # Logo max size is 1/4th of QR code size
        if logo_width > max_logo_size or logo_height > max_logo_size:
            # Use Image.Resampling.LANCZOS instead of Image.ANTIALIAS
            logo = logo.resize((max_logo_size, max_logo_size), Image.Resampling.LANCZOS)

        # Center the logo
        position = ((qr_width - logo.width) // 2, (qr_height - logo.height) // 2)
        qr_image.paste(logo, position, logo)

    # Load font
    if font_path is None:
        font_path = "/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf"  # Default font path for Linux
    if not os.path.exists(font_path):
        raise FileNotFoundError(f"Font file not found: {font_path}")

    font = ImageFont.truetype(font_path, font_size)
    draw = ImageDraw.Draw(qr_image)

    # Calculate text position
    text_bbox = draw.textbbox((0, 0), title, font=font)
    text_width = text_bbox[2] - text_bbox[0]
    text_height = text_bbox[3] - text_bbox[1]
    img_width, img_height = qr_image.size
    text_position = ((img_width - text_width) // 3, img_height - text_height - 10)  # Adjust padding as needed

    # Add title to the image
    draw.text(text_position, title, font=font, fill=fill_color)

    # Save the image
    qr_image.save(output_file)
    print(f"QR Code with title saved to {output_file}")

    return qr_image

# Define your platforms, titles, and fill colors
platforms = {
    "YouTube": ("https://www.youtube.com/@content123", "Subscribe My YouTube Channel", "blue"),
    "Facebook": ("https://www.facebook.com/content.on.demand.2024", "Follow us on Facebook", "blue"),
    "LinkedIn": ("https://www.linkedin.com/in/content-on-demand-917146269", "Connect on LinkedIn", "green"),
    "Github": ("https://github.com/Nazmul0005", "Explore our Github Repository", "purple"),
    "Instagram": ("https://www.instagram.com/contentondemand904?utm_source=qr&igsh=MTlneGt6YzIwM2R3Mw==", "Follow us on Instagram", "orange"),
    "Threads": ("https://www.threads.net/@contentondemand904", "Join us on Threads", "purple"),
    "X (Twitter)": ("https://x.com/content904?t=PaAsW4JHtSFOxTrErqrrtw&s=09", "Follow us on X (Twitter)", "cyan"),
    "TikTok": ("https://www.tiktok.com/@contentondemand?_t=8rYWYQefOPi&_r=1", "Watch us on TikTok", "DarkSlateBlue"),
}

# Example logo path (optional)
logo_path = "/content/content on demand.png"  # Set to None if you don't want to add a logo

# Generate and display QR codes
for platform, (url, title, color) in platforms.items():
    output_file = f"{platform.lower()}.png"
    qr_image = generate_qr_with_title(
        url=url,
        title=title,
        output_file=output_file,
        font_path="/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf",  # Adjust for your system
        box_size=10,
        border=4,
        fill_color=color,  # Use the unique color for each QR code
        back_color="white",
        font_size=24,
        logo=logo_path,  # Add logo if needed
        gradient=True,   # Apply gradient
    )
    print(f"Displayed QR code for {platform}.")
    display(qr_image)
    print("\n" + "-"*50 + "\n")
```
